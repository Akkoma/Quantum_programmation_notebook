2 Operators

```python
from qiskit.quantum_info import Pauli
Z0Z1 = Pauli("IZZ")
print("Z0Z1 is", Z0Z1)
print("The matrix representation of Z0Z1 is")
print(Z0Z1.to_matrix())
```

```python
print("The sparse representation of Z0Z1 is")
print(Z0Z1.to_matrix(sparse=True))
```

```python
from qiskit.quantum_info import SparsePauliOp
operator = SparsePauliOp(Pauli("IZZ")) + SparsePauliOp(Pauli("ZIZ"))
print("The operator is", operator)
print("The matrix representation of the operator is")
print(operator.to_matrix())
print("The sparse representation of the operator is")
print(operator.to_matrix(sparse=True))
```

```python
operator = SparsePauliOp(["IZI", "IXX"], coeffs = [2, 0.5])
# equivalent to:
operator = SparsePauliOp.from_list([("IZI", 2), ("IXX", 0.5)])
# equivalent to:
operator = SparsePauliOp.from_sparse_list([("Z", [1], 2), ("XX", [0,1], 0.5)],
num_qubits = 3)
print("The operator is", operator)
print("The matrix representation of the operator is")
print(operator.to_matrix())
print("The sparse representation of the operator is")
print(operator.to_matrix(sparse=True))
```

3 Expectation values

```python
from qiskit.quantum_info import Statevector, SparsePauliOp
zero = Statevector([1,0])
one = Statevector([0,1])
state = one^zero^zero
operator = SparsePauliOp("IZZ") + SparsePauliOp("ZIZ")
expectation_value = state.expectation_value(operator)
print("The exact expectation value is", expectation_value)
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp
from qiskit.primitives import Estimator
zero = Statevector([1,0])
one = Statevector([0,1])
state = one^zero^zero
qc = QuantumCircuit(3)
qc.initialize(state, range(3))
operator = SparsePauliOp("IZZ") + SparsePauliOp("ZIZ")
expectation_value = Estimator().run(circuits = qc,
observables = operator,
shots = 1024).result().values.real
print("The empirical expectation value is", expectation_value)
```
